# Instalación de Python

Por la naturaleza de nuestro curso, no utilizaremos Python de forma tradicional, sino que haremos uso de un ambiente virtual para mejorar el rendimiento de las bibliotecas que utilizaremos. De acuerdo con la documentación de [GeoCAT](https://geocat.ucar.edu/pages/software.html), se recomienda el uso de [miniconda](https://docs.conda.io/en/latest/miniconda.html), que es la versión `light` de [anaconda](https://www.anaconda.com/products/distribution).

Presentamos a continuación un paso a paso para la instalación de este software en los sistemas operativos:
- Windows
- Linux

## Instalación en Windows

### 1. Descarga de Miniconda

Puedes descargar Miniconda3 haciendo clic en el siguiente [enlace](https://docs.conda.io/en/latest/miniconda.html). Encontrarás los ejecutables para Windows, macOS y Linux. Verás una ventana con los siguientes enlaces:

![Enlace de descarga](./figs_windows/miniconda_links.png)

### 2. Abrir el archivo de instalación

Una vez descargado, abre el archivo de instalación haciendo doble clic sobre él.

### 3. Aceptar el acuerdo

Haz clic en **Next**. Debe aparecer la siguiente ventana:

![Pantalla de instalación](./figs_windows/conda2.png)

### 4. Aceptar los términos

Haz clic en **I Agree**. Luego aparecerá la siguiente ventana. Aquí selecciona **Just Me**.

![Selección Just Me](./figs_windows/conda3.png)

### 5. Configuración de la instalación

Haz clic en **Next**. Aparecerá la siguiente ventana:

![Configuración de instalación](./figs_windows/conda4.png)

En esta parte, se te indica dónde se instalará Miniconda3. Si deseas, puedes cambiar el directorio. En mi caso, lo dejaré como está.

### 6. Desmarcar la opción de PATH

Haz clic en **Next** y en la siguiente ventana, desmarca la opción **Add Miniconda3 to my PATH environment variable**. Luego, haz clic en **Install**.

![Desmarcar PATH](./figs_windows/conda5.png)

### 7. Finalización de la instalación

Una vez que termine la instalación, haz clic en **Next** y luego en **Finish**.

![Finalizar instalación](./figs_windows/conda6.png)

**¡Listo, Miniconda está instalado!**

### **Configurar el entorno de variables**

Una vez instalado Miniconda, necesitamos agregar el directorio donde se encuentra instalado.

1. En el menú de inicio de Windows, escribe **variables** y selecciona la opción **Editar las variables de entorno del sistema**.

![Editar variables](./figs_windows/edit_var.png)

2. Haz clic en **Variables de entorno**.

![Ventana de variables de entorno](./figs_windows/edit_var1.png)

3. En **Variables del sistema**, selecciona **Path** y haz clic en **Editar**. Agrega el directorio donde se encuentra instalado Miniconda. En mi caso, es el siguiente:

C:\Users\IGP\miniconda3\Scripts

4. Finalmente, haz clic en **OK** en todas las ventanas.

## Instalación en Linux

Si deseas instalar el ambiente virtual con una versión específica de Python 3, accede a la página web de Miniconda: [Miniconda](https://docs.conda.io/en/latest/miniconda.html).

Si prefieres instalar la última versión estable:

### 1. Abrir el terminal

Abre el terminal de Linux y navega a la carpeta **Downloads/Descargas**:

@user: cd ~/Downloads

*Recuerda no copiar el `@user: `, pues es solo un símbolo para representar que estamos trabajando como un usuario común en el terminal.*

### 2. Descargar Miniconda

Ejecuta el siguiente comando:

wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh

### 3. Ejecutar el instalador

Ejecuta el siguiente comando:

@user: ~/Downloads$ bash Miniconda3-latest-Linux-x86_64.sh

Verás algo como esto:

![Terminal Miniconda](./figs_linux/bash_min.png)

### 4. Aceptar el acuerdo

Presiona **ENTER** para continuar. Luego, desplázate hacia abajo con la tecla de flecha para leer el acuerdo.

Cuando llegues al final, aparecerá el siguiente mensaje:

Do you accept the license terms? [yes|no]  
[no] >>>  

Escribe `yes` y presiona **ENTER**.

### 5. Confirmar ubicación de instalación

A continuación, el instalador te pedirá la ubicación de instalación:

Miniconda3 will now be installed into this location:  
/home/georgynio/miniconda3

Presiona **ENTER** para confirmar la ubicación predeterminada o especifica una diferente si lo prefieres.

### 6. Instalar inicializador de Miniconda

El instalador te preguntará si deseas instalar el inicializador de Miniconda3:

Do you wish the installer to initialize Miniconda3 by running conda init? [yes|no]  
[no] >>> 

Escribe `yes` y presiona **ENTER**.

### 7. Finalización de la instalación

Finalmente, aparecerá un mensaje confirmando que la instalación fue exitosa.

Para actualizar tu terminal, ejecuta el siguiente comando:

@user: source ~/.bashrc

![Actualizar bash](./figs_linux/source_bash.png)
